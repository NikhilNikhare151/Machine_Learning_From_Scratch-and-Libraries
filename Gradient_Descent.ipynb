# Gradient Descent for Linear Regression

```python
import numpy as np
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
```

```python
x,y= make_regression(n_features=1,n_samples=40,noise=11)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(7,4))
plt.scatter(x,y)
plt.show()
```

```python
epochs=100
lr=0.007
```

```python
m= 0.02
c=30.43
```

```python
error=[]
```

```python
plt.scatter(x,y)
for i in range(epochs):
    gd_m= -2 * np.sum((y - m * x.ravel() - c) * x.ravel())
    gd_c= -2 * np.sum(m * x.ravel() - c)

    y_pred= m * x.ravel() + c
    error.append(np.sum(np.square(y-y_pred)))

    
    plt.plot(x,y_pred)
    
    
    m= m - (lr * gd_m)
    c= c - (lr * gd_c)

plt.show()
```
