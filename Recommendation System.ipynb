# Content Based Analysis

```python
user_matrix=pd.pivot_table(dummy, index="user_id", columns="title", values="rating", fill_value='0')
user_matrix
```

```python
movies_corr=user_matrix.corr()
movies_corr
```

```python
movies_corr["Star Wars (1977)"].sort_values(ascending=False).head(6)[1:]
```

```python
def movie_recommender(movie):
    return movies_corr[movie].sort_values(ascending=False).head(6)[1:]
```

```python
movie_recommender("Event Horizon (1997)")
```

```python
movie_recommender("Toy Story (1995)")
```

```python
movie_recommender("Grosse Pointe Blank (1997)")
```

```python
movie_recommender("Event Horizon (1997)").index
```

```python
dummy[dummy["title"].isin(movie_recommender("Event Horizon (1997)").index)]
```

```python
dummy[(dummy["title"].isin(movie_recommender("Event Horizon (1997)").index)) &
        (- dummy["title"].isin(["Event Horizon (1997)"]))]["user_id"].tolist()
```

```python
def user_id_list(movie_name):
    return set(dummy[(dummy["title"].isin(movie_recommender(movie_name).index)) &
        (- dummy["title"].isin([movie_name]))]["user_id"].tolist())
```

```python
len(user_id_list("Event Horizon (1997)"))
```

# Collebrative Filtering

```python
movie_matrix = pd.pivot_table(dummy, index="title", columns="user_id", values="rating", fill_value='0')
movie_matrix
```

```python
user_corr = movie_matrix.corr()
user_corr
```

```python
user_corr[123].sort_values(ascending=False).head(6)[1:]
```

```python
# find movies seen by above user id but user 123 dosent seen
```

```python
similar_user = user_corr[123].sort_values(ascending=False).head(6)[1:].index
similar_user
```

```python
def similar_user(user_id):
   return user_corr[user_id].sort_values(ascending=False).head(6)[1:].index 
```

```python
similar_user(123)
```

```python
def movie_list(user_id):
    return dummy[(dummy["user_id"].isin(similar_user(user_id))) & 
    (dummy["user_id"]!=123)]["title"].value_counts().head(5).index.tolist()
```

```python
movie_list(123)
```
